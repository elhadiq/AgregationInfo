```python
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,Y=np.array([0,0.5,1]),np.array([1/5,2,1/4])
N=len(X)
```

```python
def computeParames(X,Y,D,N):
    A,C,E,F=[],[],[],[]
    for n in range(1,N):
        A.append((X[n]-X[n-1])/(X[-1]-X[0]))
        E.append(X[n-1]-A[n-1]*X[0])
        C.append((Y[n]-Y[n-1]-D[n-1]*(Y[-1]-Y[0]))/(X[-1]-X[0]))
        F.append(Y[n-1]-C[n-1]*X[0]-D[n-1]*Y[0])
    return A,C,E,F
```

```python
wx=lambda n,x:A[n]*np.array(x)+E[n]
wy=lambda n,x,y:C[n]*np.array(x)+D[n]*np.array(y)+F[n]
```

```python
Ax=np.linspace(-1,1,10)
Ay=np.linspace(-2,3,10)
```

```python
def iterer(Ax,Ay,iteration=1):
    if iteration==0:
        return Ax,Ay
    AAx,AAy=[],[]
    for n in range(N-1):
        AAx.extend(wx(n,Ax))
        AAy.extend(wy(n,Ax,Ay))
    return iterer(AAx,AAy,iteration-1)
```

```python
D=np.array([0.33, 0.076])
A,C,E,F=computeParames(X,Y,D,N)
x,y=iterer(Ax,Ay,10)
plt.plot(x,y)
```

```python
D=np.array([0.7,0.9])
A,C,E,F=computeParames(X,Y,D,N)
x,y=iterer(Ax,Ay,10)
plt.plot(x,y)
```
