```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
epsilon= lambda a,t:np.exp(-a**2/(a**2-t**2)) if abs(t)<a else 0
x=np.linspace(-10,10,1000)
y=list(map(lambda z:epsilon(2,z),x))
plt.plot(x,y)
```
